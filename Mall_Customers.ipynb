```python
import pandas as pd
```

```python
path = '~/Downloads/Mall_Customers.csv'
df = pd.read_csv(path)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], bins=15, kde=True, color='purple')
plt.title('Distribusi Umur Pelanggan')
plt.xlabel('Umur')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', s=100)
plt.title('Pendapatan vs Skor Belanja')
plt.grid(True)
plt.show()
```

```python
# Menghitung rata-rata skor belanja berdasarkan gender
df_gender = df.groupby('Gender')['Spending Score (1-100)'].mean()
print(df_gender)

# Visualisasi
df_gender.plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'lightblue'], startangle=90)
plt.title('Rata-rata Skor Belanja: Pria vs Wanita')
plt.ylabel('')
plt.show()
```
